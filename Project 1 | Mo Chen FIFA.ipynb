# Data Cleaning

Changing Height & Weight units
 


Weight (KG to LBS)

Height (cm to feet)

```python
players_copy = players.copy()
players_copy
```

```python
#MUST REFRESH THE DEMO before running this section of CODE!!!

lbslist = []
for weight in players_copy['Weight']:
    weight = weight * 2.2046
    weight = round(weight,0)
    lbslist.append(weight)
    
players_copy['Weight'] = lbslist
```

```python
#MUST REFRESH THE DEMO before running this section of CODE!!!

ftlist = []
for height in players_copy['Height']:
    height*= 0.0328
    height = round(height,1)
    ftlist.append(height)

players_copy['Height'] = ftlist
```

```python
players_copy = players_copy.rename({'Height':'Height (ft)','Weight':'Weight (lbs)'},axis = 'columns')
players_copy.head()
```

```python
players_copy.dropna(how = 'all')
players_copy.ContractUntil = players_copy['ContractUntil'].astype(int)
```
