# Data Cleaning


Weight (kg --> lbs)

Height (cm --> feet)

```python
players_copy = players.copy()
players_copy
```

```python
#MUST REFRESH THE dataset COPY code before running this section of CODE!!!

lbslist = []
for weight in players_copy['Weight']:
    weight = weight * 2.2046
    weight = round(weight,0)
    weight = f"{weight} lbs"
    lbslist.append(weight)
    
players_copy['Weight'] = lbslist
```

```python
#MUST REFRESH THE dataset COPY code before running this section of CODE!!!

ftlist = []
for height in players_copy['Height']:
    inches = height/2.54
    feet = inches/12
    ft_height = int(feet)
    in_height = round(inches - (ft_height*12))
    if ((ft_height == 5) and (in_height == 12)):
        ft_height = 6
        in_height = 0
    height = f"{ft_height}' {in_height}"
    ftlist.append(height)

players_copy['Height'] = ftlist
```

Renaming Height and Weight column names

```python
players_copy = players_copy.rename({'Height':'Height (ft)','Weight':'Weight (lbs)'},axis = 'columns')
players_copy.head()
```

Player Value and Release Clause unit transformation

```python
#VALUE UNIT TRANSFORMATION
#MUST REFRESH THE dataset COPY code before running this section of CODE!!!

millionlist = []
for value in players_copy['ValueEUR']:
    value = value/1000000
    value = f"{value} mil"
    millionlist.append(value)
    
players_copy['ValueEUR'] = millionlist
```

```python
players_copy['ValueEUR'].head()
```

```python
#RELEASE CLAUSE TRANSFORMATION
#MUST REFRESH THE dataset COPY code before running this section of CODE!!!

releaseclause = []
for clause in players_copy['ReleaseClause']:
    if clause > 999999999:
        clause = round(clause / 100000000,1)
        clause = f"{clause} bil" 
    else:
        clause = round(clause/1000000,1)
        clause = f"{clause} mil"
    releaseclause.append(clause)

players_copy['ReleaseClause'] = releaseclause
```
